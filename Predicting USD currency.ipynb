# Currency predictions with Machine learning

Using **pandas**, **matplotlib** and **sklearn**

All we need - a .xlsx file containing currency prices.
What we want to do - take the prevoius 28 days and make a prediction about the next 7.

```python
import pandas as pd
money = pd.read_excel("USD_CURRENCY.xlsx")
money.describe() # shows the statistics of the given dataset
```

```python
money # that's our dataset 
```

```python
future = 7
past = 28
```

```python
import sklearn
values = money.curs # keeping all the values form money.curs into a seperate array
start = past
end = len(values) - future
```

```python
values
```

```python
raw_data = [] # keeping all the information needed to to make the predictions - the previous values and the next which will be used to test our ML model
for i in range(start, end):
    pfv = values[(i - past):(i + future)]
    raw_data.append(list(pfv))
```

```python
raw_data
```

```python
past_columns = [] # keeping the previous values
for i in range(past):
    past_columns.append("past_{}".format(i))
future_columns = [] # keeping the next values
for i in range(future):
    future_columns.append("future_{}".format(i))
```

```python
df = pd.DataFrame(raw_data, columns = (past_columns + future_columns))
```

```python
X = df[past_columns][:-1]
y = df[future_columns][:-1]
X_test = df[past_columns][-1:]
y_test = df[future_columns][-1:]
```

We've prepared our final datasets: X, y, X_test, y_test

So now, we're ready to make our ML model.

# 1. Linear Regression

```python
from sklearn.linear_model import LinearRegression
LinReg = LinearRegression()
LinReg.fit(X, y)
prediction = LinReg.predict(X_test)[0]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(prediction, label = "Prediction")
plt.plot(y_test.iloc[0], label = "Real")
plt.legend()
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
mean_absolute_error(y_test, [prediction])
```

## Linear Regression - result result for 7 days using last 28

The mean_absolute_error is __0.8024480001128244__

# 2. KNeighborsRegressor

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
KNN = KNeighborsRegressor(n_neighbors = 1)
KNN.fit(X, y)
```

```python
prediction = KNN.predict(X_test)[0]
mean_absolute_error(y_test, [prediction])
```

```python
plt.plot(prediction, label = "Prediction")
plt.plot(y_test.iloc[0], label = "Real")
plt.legend()
```

## K Neighbors Regression - result for 7 days using last 28

The mean_absolute_error is __0.17897142857142864__

```python
future = 100
past = 300
values = money.curs
start = past
end = len(values) - future
raw_data = []
for i in range(start, end):
    pfv = values[(i - past):(i + future)]
    raw_data.append(list(pfv))
past_columns = []
for i in range(past):
    past_columns.append("past_{}".format(i))
future_columns = []
for i in range(future):
    future_columns.append("future_{}".format(i))
df = pd.DataFrame(raw_data, columns = (past_columns + future_columns))
X = df[past_columns][:-1]
y = df[future_columns][:-1]
X_test = df[past_columns][-1:]
y_test = df[future_columns][-1:]
KNN = KNeighborsRegressor(n_neighbors = 1)
KNN.fit(X, y)
```

```python
prediction = KNN.predict(X_test)[0]
mean_absolute_error(y_test, [prediction])
```

```python
plt.plot(prediction, label = "Prediction")
plt.plot(y_test.iloc[0], label = "Real")
plt.legend()
```
